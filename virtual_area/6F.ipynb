```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,7))




pose_ax = fig.add_subplot(1, 1, 1, projection='3d')
pose_ax.view_init(29,-136)
pose_ax.set_xlim3d(0, 1030*1)
pose_ax.set_ylim3d(0, 1114*1)
pose_ax.set_zlim3d(0, 1000*1)

a=[1,1,1]
# x, y, z 
xpos = [0,720,918]
ypos = [562,477,248]
zpos = [0,0,0]

transxpos= [a[i]*xpos[i] for i in range(len(a))]
transypos= [a[i]*ypos[i] for i in range(len(a))]
transzpos= [a[i]*zpos[i] for i in range(len(a))]

# delta x, y, z 
dx = [407,198,90]
dy = [553,630,866]
dz = [400,400,400]
transdx= [a[i]*dx[i] for i in range(len(a))]
transdy= [a[i]*dy[i] for i in range(len(a))]
transdz= [a[i]*dz[i] for i in range(len(a))]

# 
pose_ax.bar3d(transxpos,transypos,transzpos,transdx,transdy,transdz,'#7f7f7f',alpha=0.2) #,color='#00ceaa'()  

pose_ax.scatter(0, 0,0,color='green')
pose_ax.text(0, 0,0, "(0,0,0) starting point", color='red')

pose_ax.scatter(0, 562,0,color='green')
pose_ax.text(0, 562,0, "(0,562,0)", color='red')

pose_ax.scatter(407, 1115,400,color='green')
pose_ax.text(407, 1115,400, "(407, 1115,400)", color='red')

pose_ax.scatter(720, 477,0,color='green')
pose_ax.text(720, 477,0, "(720, 477,0)", color='red')

pose_ax.scatter(918, 1107,400,color='green')
pose_ax.text(918, 1107,400, "(918, 1107,400)", color='red')


pose_ax.scatter(918,248,0,color='green')
pose_ax.text(918,248, 0, "(918,248,0)", color='red')

pose_ax.scatter(1008,1114,400,color='green')
pose_ax.text(1008,1114,400, "(1008,1114,400)", color='red')





pose_ax.set_title('6F')
pose_ax.set_xlabel('x axis')
pose_ax.set_ylabel('y axis')
pose_ax.set_zlabel('z axis')
fig.tight_layout()
plt.show()
```
